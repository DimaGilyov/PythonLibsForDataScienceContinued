## Урок 4. Оценка и интерпретация полученной модели

#### 1) Расскажите, как работает регуляризация в решающих деревьях, какие параметры мы штрафуем в данных алгоритмах?

В узком смысле Регуляризацию (обычно определяемую как добавление штрафа за гребень или Лассо) трудно реализовать для Деревьев. Дерево-это эвристический алгоритм.

В более широком смысле Регуляризация (как любое средство предотвращения перенапряжения) деревьев осуществляется путем:

* предел макс. глубина деревьев
* ансамбли / сумка больше, чем просто 1 дерево
* установите более строгий критерий остановки для дальнейшего разделения узла (например, минимальное усиление, количество выборок и т.д.).

#### 2) По какому принципу рассчитывается "важность признака (feature_importance)" в ансамблях деревьев?

Как правило, важность обеспечивает оценку, которая указывает, насколько полезной или полезной была каждая функция при построении расширенных деревьев решений в модели. Чем больше атрибут используется для принятия ключевых решений с деревьями решений, тем выше его относительная важность.

Эта важность вычисляется явно для каждого атрибута в наборе данных, что позволяет ранжировать атрибуты и сравнивать их друг с другом.

Важность рассчитывается для отдельного дерева решений по величине, на которую каждая точка разделения атрибута улучшает показатель производительности, взвешенный по количеству наблюдений, за которые отвечает узел. Показателем производительности может быть чистота (индекс Джини), используемая для выбора точек разделения, или другая более конкретная функция ошибок.

Значения характеристик затем усредняются по всем деревьям в модели.
